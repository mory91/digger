```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
r = json.load(open('../result.json', 'r'))
```

```python
times = r['times']
result = {}
for met in ['mae', 'mse', 'smape']:
    if met not in result:
        result[met] = []
    result[met] = [rr[met] for rr in r['evals']]
```

```python
r
```

```python
for metric in result:
    plt.ylabel(metric)
    test_scores = result[metric]
    plt.title(f"{metric} Score of different time deltas")
    plt.plot(times, test_scores, 'b-', label='test score')
    plt.legend(loc='best')
    plt.show()
```
